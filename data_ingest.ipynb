```python
def read_organize_data():
    """
    Fill this function out, should return a dataframe with image object as 3D array, and category encoding 0-19
    #https://stackoverflow.com/questions/15312953/choose-a-file-starting-with-a-given-string
    #https://stackoverflow.com/questions/28339000/unzip-zip-files-in-folders-and-subfolders-with-python/30591949
    #https://stackoverflow.com/questions/36285502/how-to-extract-zip-file-recursively-in-python
    """

    #Get filepath
    file_path = os.path.join(os.getcwd(), 'drive/My Drive/D100_Final_Project/20_categories_training')
    #Empty df w image and label
    df = pd.DataFrame(columns=['image', 'label'])
    #Get list of folders
    folders = os.listdir(file_path)
    #Alpha sort list of folders
    folders.sort()

    i = 0 #i = encoding number
    for file_name in folders: #for each subfolder
        if file_name.startswith('.'): #check image name format ._
            continue
        print(file_name) #checkpoint

        #For each image
        for image_file in os.listdir(os.path.join(file_path, file_name)):
            if image_file.startswith('.'): #check image name format ._
                continue
            #get entire filepath
            image_file = os.path.join(file_path, file_name, image_file) 
            #read image with io
            image = io.imread(image_file)
            # skimage.io.imshow(image)
            # plt.show()
            df = df.append({'image': image, 'label': i}, ignore_index=True)
        i += 1
    return df

```
